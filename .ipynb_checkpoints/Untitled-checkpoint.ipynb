```python
# imports

import pandas as pd 
import numpy as np
from datetime import datetime, timedelta
from sklearn.decomposition import PCA
from sklearn import preprocessing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#bringing in our 4 dataframes 
public_data = pd.read_csv("COVID-19_Aggregate_Cases_Current_Weekly_County_Health.csv")
#google_data_2020 = pd.read_csv("2020_sub_region_1_daily_2020_US_Pennsylvania_daily_symptoms_dataset.csv",  dtype=str)
google_data_2021 = pd.read_csv("2021_sub_region_1_daily_2021_US_Pennsylvania_daily_symptoms_dataset(1).csv", dtype=str)
google_data_2022 = pd.read_csv("2022_sub_region_1_daily_2022_US_Pennsylvania_daily_symptoms_dataset(1).csv", dtype=str)
```

```python
# consolidating our google data into one dataframe
google_dataframes = [google_data_2021, google_data_2022]
google_data = pd.concat(google_dataframes)
google_data.head()
```

```python
#removing extraneous columns from our `google_data`
google_data.drop(columns=["country_region_code", "country_region", "sub_region_1",
                          "sub_region_1_code", "place_id"], inplace=True)

google_data.head()
```

```python
# make the date our index, then sort by the date index
google_data['date']= pd.to_datetime(google_data['date'])
google_data.set_index('date', inplace=True)
google_data = google_data.sort_index()
```

```python
google_data.head()
```

```python

google_col = google_data.columns.to_list()

#symptom_col = google_col[2:]
```

```python
#changing our symptoms columns dtype to float
google_data[google_col[2:]] = google_data[google_col[2:]].astype(float)
```

```python
# checking for null values
google_data.isnull().sum()
```

```python
# start processing the public_data
public_data['Date'] = pd.to_datetime(public_data['Date'])
public_data.set_index('Date', inplace=True)
public_data = public_data.sort_index()
```

```python
public_data.head()
```

```python
public_philly = public_data[public_data.Jurisdiction == 'Philadelphia']
#making it so the dates will match when we join the two dataframes
public_philly = public_philly.loc["2021-01-01" : "2022-11-13"]
public_philly
```

```python
# no null values to deal with here
public_philly.isnull().sum().sum()
```

```python
# drop columns that aren't needed, 
#removing extraneous columns from our `google_data`
public_philly.drop(columns=["Jurisdiction", "Population (2019)", "County FIPS Code",
                          "Longitude", "Latitude", "Georeferenced Lat & Long"], inplace=True)

public_philly.head()
```

```python
#join data sets and decide on target variable 
philly_data = pd.concat([google_philly, public_philly], axis= 1)
target = philly_data['New Cases']
symptoms  = philly_data[symptom_list]
```
